# Passo 2: Visualizar base de dados

```python

tabela = pd.read_csv("acidente-transito.csv", sep=';', encoding='latin-1')


display(tabela)
```

# Passo 3: Corrigir erros da base de dados
#Corrigindo valores vazios

#Excluindo linhas vazios

#Excluindo tabelas desnecessaria 

```python

# Visualizar informações antes das operações
display(tabela.info())

#Remover espaços em branco dos nomes das colunas
tabela.columns = tabela.columns.str.strip()

# Remover linhas com valores ausentes e colunas específicas
tabela = tabela.dropna()
tabela = tabela.drop(columns=["Nº_boletim", "seq_veic"])

# Visualizar informações após as operações
display(tabela.info())
```

# Passo 4: Análise inicial dos dados
Que tipo de socorro ocorreu, qual foi o veiculo que mais ocorreu acidente,Situação em que o veiculo estava, horario que mais ocorreu acidente.

```python
# Que tipo de socorro ocorreu
tipo_socorro_mais_comum = tabela['desc_tipo_socorro'].mode().values[0]
print("Tipo de socorro mais comum:", tipo_socorro_mais_comum)

# Veículo que mais ocorreu acidente
veiculo_mais_comum = tabela['descricao_especie'].mode().values[0]
print("Veículo que mais ocorreu acidente:", veiculo_mais_comum)

# Situação em que o veículo estava
situacao_veiculo_mais_comum = tabela['desc_situacao'].mode().values[0]
print("Situação em que o veículo estava:", situacao_veiculo_mais_comum)

# Horário que mais ocorreu acidente
horario_mais_ocorrido = tabela['data_hora_boletim'].str.split().str[1].str.split(':').str[0].astype(int).mode().values[0]
print("Horário que mais ocorreu acidente:", horario_mais_ocorrido, "horas")

```

# Passo 4: Análise 
Agora mostrando em graficos os resultado da pesquisa anterior

```python


# Dados para o gráfico
tipos_socorro = tabela['desc_tipo_socorro'].value_counts()
veiculos = tabela['descricao_categoria'].value_counts()
situacoes_veiculo = tabela['desc_situacao'].value_counts()

# Plotar gráficos de barras
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

axs[0, 0].bar(tipos_socorro.index, tipos_socorro.values, color='skyblue')
axs[0, 0].set_title('Tipos de Socorro')
axs[0, 0].set_ylabel('Quantidade')
axs[0, 0].tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo x

axs[0, 1].bar(veiculos.index, veiculos.values, color='salmon')
axs[0, 1].set_title('Veículos Envolvidos')
axs[0, 1].set_ylabel('Quantidade')
axs[0, 1].tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo x

axs[1, 0].bar(situacoes_veiculo.index, situacoes_veiculo.values, color='lightgreen')
axs[1, 0].set_title('Situação dos Veículos')
axs[1, 0].set_ylabel('Quantidade')
axs[1, 0].tick_params(axis='x', rotation=45)  # Rotaciona os rótulos do eixo x

# Remover o último subplot
fig.delaxes(axs[1,1])

plt.tight_layout()
plt.show()
```

# Passo 4: Análise 
Mostrando que veiculos teve mais acidente a especificação dele.
Este código está gerando um gráfico de pizza mostrando a distribuição percentual dos tipos de veículos envolvidos nos acidentes. Cada fatia do gráfico representará um tipo de veículo e a porcentagem de ocorrência desse tipo em relação ao total.

```python


# Calcular a contagem de 'descricao_especie'
especie_count = tabela['descricao_especie'].value_counts()

# Calcular as porcentagens
especie_perc = 100. * especie_count / especie_count.sum()

# Filtrar as 'descricao_especie' com mais de 1.8% de ocorrências
especie_filtrado = especie_perc[especie_perc >=1.7  ]

# Gerar o gráfico de pizza
fig, ax = plt.subplots(figsize=(10, 10))
wedges, texts, autotexts = ax.pie(especie_filtrado, autopct='%1.1f%%', startangle=140)

# Mover a legenda para fora do gráfico e rotacioná-la
ax.legend(wedges, especie_filtrado.index,
          title="descricao_especie",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(texts, rotation_mode="anchor", ha="right")

plt.title('Distribuição de descricao_especie (acima de 1.7%)')
plt.show()
```

# Passo 4: Análise mais detalhada nos mesês e nos tipos de veiculos. Quais veiculos teve mais acidente em qual mes

```python


# Converter 'data_hora_boletim' para datetime
tabela['data_hora_boletim'] = pd.to_datetime(tabela['data_hora_boletim'])

# Extrair a hora
tabela['hora'] = tabela['data_hora_boletim'].dt.hour

# Agrupar por hora, 'descricao_especie' e 'desc_situacao'
grupo = tabela.groupby(['hora', 'descricao_especie', 'desc_situacao']).size()

# Encontrar a combinação com a contagem máxima
hora_max, especie_max, situacao_max = grupo.idxmax()
print(f"'descricao_especie' e 'desc_situacao' com o maior número de ocorrências é:{especie_max}e {situacao_max}")


```

```python
# Converter 'data_hora_boletim' para datetime
tabela['data_hora_boletim'] = pd.to_datetime(tabela['data_hora_boletim'], format='%d/%m/%Y %H:%M')

# Extrair o mês
tabela['mes'] = tabela['data_hora_boletim'].dt.month

# Agrupar por mês
grupo_mes = tabela.groupby('mes').size()

# Encontrar o mês com a contagem máxima
mes_max = grupo_mes.idxmax()
print(f"O mês com o maior número de ocorrências é: {mes_max}")

# Agrupar por mês e 'descricao_especie'
grupo_mes_especie = tabela.groupby(['mes', 'descricao_especie']).size()

# Encontrar a 'descricao_especie' com a contagem máxima para o mês específico
especie_max_mes = grupo_mes_especie[mes_max].idxmax()
print(f"A 'descricao_especie' com o maior número de ocorrências no mês {mes_max} é: {especie_max_mes}")


```

```python

# Obter a lista de 'descricao_especie' únicas
especies = tabela['descricao_especie'].unique()

# Para cada 'descricao_especie', criar um gráfico de barras
for especie in especies:
    # Filtrar os dados para a 'descricao_especie' atual
    dados_especie = tabela[tabela['descricao_especie'] == especie]
    
    # Agrupar por mês e contar o número de ocorrências
    contagem_por_mes = dados_especie['mes'].value_counts().sort_index()
    
    # Criar o gráfico de barras
    plt.figure(figsize=(12, 6))
    plt.bar(contagem_por_mes.index, contagem_por_mes)
    plt.xlabel('Mês')
    plt.ylabel('Número de Ocorrências')
    plt.title(f"Número de Ocorrências da '{especie}' por Mês")
    plt.show()
```
